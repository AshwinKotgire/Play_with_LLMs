##Sources :

2. FAISS : https://python.langchain.com/docs/integrations/vectorstores/faiss
3. CharacterTextSplitter : https://python.langchain.com/docs/modules/data_connection/document_transformers/text_splitters/character_text_splitter


```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import FAISS
from langchain.document_loaders import TextLoader
from langchain.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader("Chanakyachant.pdf")

docs = loader.load_and_split()
# OR
```

```python
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)
```

```python
len(docs)
```

HuggingFace Embeddings :
1. Blog : https://huggingface.co/blog/getting-started-with-embeddings
2. Langchain : https://python.langchain.com/docs/integrations/providers/huggingface


```python
embeddings = HuggingFaceEmbeddings()
```

```python
db = FAISS.from_documents(docs, embeddings)
```

For multiple queries use embeddings.embed_document()

```python
result=db.similarity_search('Where did the plane crash', k=5)
```

```python
result[0]
```

##Extra Observations:

2. installing FAISS with correct device is must( faiss-cpu or faiss-gpu)
